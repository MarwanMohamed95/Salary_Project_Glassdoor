```python
df_cat = df[['Sector','hourly', 'Same State', 'Python', 'Spark', 'Excel', 'AWS', 'Job_simp', 'Seniority']]
```

```python
for i in df_cat.columns:
    x = df_cat[i].value_counts()
    print("graph for %s ,Total %s"%(i,len(x)))
    sns.barplot(x = x.index,y = x.values)
    plt.xticks(rotation=90)
    plt.show()
```
